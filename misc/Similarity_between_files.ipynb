## Similarity between documents (TFIDF, cosine similarity)

How to describe the similarity between two text documents


```python
import re
from collections import defaultdict
```

```python
files = []
for i in range(1,5):
    with open('../data/file%d.txt' % i, 'r') as f:
        fl = []
        for line in f:
            line = re.sub(r'[^a-zA-Z0-9\' ]', '', line)
            line = line.strip().lower()
            fl.append(line)
        files.append(' '.join(fl))
        
print files[0][:500]
print files[1][:500]
```

```python
aw = set()
tfl = []      # term frequency list

for doc in files:
    d = defaultdict(int)
    wordlist = doc.split(' ')
    for w in wordlist:
        d[w] += 1
        aw.add(w)
    tfl.append(d)

al = list(aw)
print al[:100]
print len(al)
```

```python
import math

idf = []
n = len(files)

for w in al:
    nw = 0
    for tf in tfl:
        if w in tf:
            nw+=1
    idf.append(math.log(n*1.0/nw))
print idf[:100]
```

```python
tfidfl = []
for i in range(n):
    tf = tfl[i]
    tfidf = [0.0 for x in range(len(al))]
    for w in tf:
        ind = al.index(w)
        tfidf[ind] = tf[w] * idf[ind]
    tfidfl.append(tfidf)
    
print tfidfl
```

```python
def cosine_similarity(list1, list2):
    sum = 0.
    norm1 = 0.
    norm2 = 0.
    if len(list1) != len(list2):
        return None
    for i in range(len(list1)):
        sum += list1[i] * list2[i]
        norm1 += list1[i] * list1[i]
        norm2 += list2[i] * list2[i]
    return sum/math.sqrt(norm1)/math.sqrt(norm2)
```

```python
maxind=0
maxscore=0
scores=[]
for i in range(n):
    score = cosine_similarity(tfidfl[0], tfidfl[i])
    scores.append(score)
print scores
print max(scores)
print scores.index(max(scores))+2
```

## Compact solution

```python
import math

files = []
for i in range(1,5):
    with open('../data/file%d.txt' % i, 'r') as f:
        fl = []
        for line in f:
            line = re.sub(r'[^a-zA-Z0-9\' ]', '', line)
            line = line.strip().lower()
            fl.append(line)
        files.append(' '.join(fl))

count = lambda word, doc: sum(w == word for w in doc.split())
inner = lambda a, b: [aa * bb for aa, bb in zip(a, b)]
dot = lambda a, b: sum(inner(a, b))

alltexts = ' '.join(files)
words = list(set(alltexts.split()))
tf = [[count(word, doc) for word in words] for doc in files]
idf = [math.log(1./(sum(word in doc for doc in files))) for word in words]
tfidf = [inner(t, idf) for t in tf]
cosines = [dot(tfidf[0], t) / math.sqrt(dot(tfidf[0], tfidf[0])) / math.sqrt(dot(t, t))
           for t in tfidf[1:]]

# We start with the second document
print cosines.index(max(cosines)) + 2
```
